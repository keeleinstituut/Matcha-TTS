# Separating Corpus WAV and TXT Files, and Renumbering Sentences
This notebook separates the WAV and TXT files from corpora into dedicated folders and optionally renumbers the files. Separating is necessary for following Matcha-TTS input format and renumbering not to overwrite any data in the later steps.

```python
import os
import shutil
```

```python
# Paths to input corpora (each folder contains both WAV and TXT files)
korp_1 = '/home/sandra/projects/matcha/data/peeter/peeter_jutustav'
korp_2 = '/home/sandra/projects/matcha/data/peeter/peeter_otsekone'

# Output folders for separated and renumbered WAV and TXT files
wav_path = '/home/sandra/projects/matcha/data/peeter/peeter_wav'
txt_path = '/home/sandra/projects/matcha/data/peeter/peeter_txt'
```

```python
def pad_no(number, padding=5):
    """Pad the number with leading zeros to the specified width."""
    return str(number).rjust(padding, '0')

def rearrange_files(original_path, wav_output_path, txt_output_path, renumber=False, start_numbering_from=1, numbering_padding_len=5, prefix='lause'):
    """Copy and optionally renumber WAV and TXT files from the original corpus folder to separate output folders.
    
    Args:
        original_path (str): Path to the original corpus folder containing WAV and TXT files.
        wav_output_path (str): Output folder for WAV files.
        txt_output_path (str): Output folder for TXT files.
        renumber (bool): If True, renumber sentences starting from 'start_numbering_from'.
        start_numbering_from (int): The starting number for renumbering (if renumber=True).
        numbering_padding_len (int): Number of digits for zero-padding the numbering.
        prefix (str): Prefix for the new file names when renumbering.
    """
    counter = start_numbering_from

    print(len(os.listdir(original_path)))
    for fname in os.listdir(original_path):
        type = fname.split('.')[-1]
        name = fname.split('.')[0]

        if type != 'wav':
            continue

        if renumber:
            new_name = prefix + pad_no(counter, numbering_padding_len)
            counter += 1
        else:
            new_name = name

        # Copy the WAV file to the output folder
        shutil.copyfile(os.path.join(original_path, fname), os.path.join(wav_output_path, new_name + '.' + type))

        # Copy the corresponding TXT file to the output folder
        shutil.copyfile(os.path.join(original_path, name + '.txt'), os.path.join(txt_output_path, new_name + '.txt'))
```

```python
# Add more function calls below if you have additional corpora to process.
# Make sure to set 'renumber=True' and 'start_numbering_from' correctly for each corpus that needs renumbering. This value must be checked and set manually for each new corpus.
rearrange_files(korp_1, wav_path, txt_path, renumber=False)
rearrange_files(korp_2, wav_path, txt_path, renumber=True, start_numbering_from=3376, numbering_padding_len=5)
```
