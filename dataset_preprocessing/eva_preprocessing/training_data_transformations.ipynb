```python
from sklearn.model_selection import train_test_split
```

### 1. Jaga andmed splittidesse

```python
txt_folder = 'path/matcha/eva_data/eva_full_training_set.txt'

training_files_folder = 'path/matcha/eva_training_sets/eva_exp_ms_1' # Output folder (txt-files with file path and sentences)
```

```python
with open(txt_folder, 'r', encoding='utf-8') as f:
    training_lines = f.readlines()
```

```python
train, test_val = train_test_split(training_lines, train_size=0.95, shuffle=True, random_state=5)

val, test = train_test_split(test_val, train_size=0.1, shuffle=True, random_state=5)
```

### 2. Modifitseeri andmeid vastavalt kõnelejale

1. yksiksonad_1 ja yksiklaused speaker ID = 0
2. yksiksonad_2 speaker ID = 1

```python
def create_multi_speaker_lines(split):
    new_split = []
    for row in split:
        if 'yksiksonad_2' in row:
            speaker_id = 1
        else:
            speaker_id = 0
        path, text = row.split('|')
        new_line = f"{path.strip()}|{speaker_id}|{text.strip()}\n"
        new_split.append(new_line)
    return new_split

train = create_multi_speaker_lines(train)
test = create_multi_speaker_lines(test)
val = create_multi_speaker_lines(val)
```

### 3. Andmete õigesse vormingusse viimine

```python
train_filelist_fname = 'est_audio_text_train_filelist.txt'
validation_filelist_fname = 'est_audio_text_val_filelist.txt'
test_filelist_fname = 'est_audio_text_test_filelist.txt'
```

```python
def write_to_filelist(filelist_name, set):    
    with open(filelist_name, 'w') as f:
        f.writelines(set)
```

```python
# Process train data
write_to_filelist(f'{training_files_folder}/{train_filelist_fname}', train)

# Process validation data
write_to_filelist(f'{training_files_folder}/{validation_filelist_fname}', val)

# Process test data
write_to_filelist(f'{training_files_folder}/{test_filelist_fname}', test)
```
