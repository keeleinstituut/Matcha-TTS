Üksiksõnade tekstide eraldamine üksikutesse failidesse.

```python
import os

def extract_txt_files_to_folder(target_folder, ndx_file):
    """
    ndx-failist tekstifailide genereerimine ja sihtkausta salvestamine
    """

    # Sihtkausta puudumisel loo sihtkaust
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)

    # Eralda ndx-faili sisu eraldi failidesse
    for line in open(ndx_file):
        fname, transcript = line.split('\t')
        if '+' in fname: # üksiklausete failide puhul
            fname = fname.replace('+', '_')
        new_fname = fname.strip().replace('.wav', '.txt')
        with open(os.path.join(target_folder, new_fname), 'w', encoding='utf-8') as f:
            f.write(transcript.strip())
```

```python
# yksiksonad_1 
ndx_path = '/path/matcha/eva_data/yksiksonad_1/eva_yksiksonad1_toimetatud.ndx'
target_path = '/path/matcha/eva_data/yksiksonad_1/yksiksonad_1_txt'
extract_txt_files_to_folder(target_path, ndx_path)
```

/path/matcha/data/peeter_wav_resampled/lause02178.wav|Ma tunnistan teile, et mul oli kogu aja, juba oma jutustuse algusest peale hirm selle hetke ees.

```python
def create_training_set_row(text_folder, wav_path):
    """
    Tekstifailide ja wav-failide paaride genereerimine
    """
    for filename in os.listdir(text_folder):
        with open(os.path.join(text_folder, filename), 'r', encoding='utf-8') as f:
            text = f.read()
            wav_file = os.path.join(wav_path, filename.replace('.txt', '.wav'))
            if not os.path.isfile(wav_file):
                print(f'Wav file not found: {wav_file}')
                continue
        yield f'{wav_file}|{text}'
```

```python
full_training_set_path = '/path/matcha/eva_data/eva_full_training_set.txt'

with open(full_training_set_path, 'w', encoding='utf-8') as f:
    yksiksonad_1_txt_path = '/path/matcha/eva_data/yksiksonad_1/yksiksonad_1_txt'
    for line in create_training_set_row(yksiksonad_1_txt_path, '/home/sandra/Documents/matcha/eva_data/yksiksonad_1/yksiksonad_1_wav_resampled'):
        f.write(line + '\n')
    
    yksiksonad_2_txt_path = '/path/matcha/eva_data/yksiksonad_2/yksiksonad_2_txt'
    for line in create_training_set_row(yksiksonad_2_txt_path, '/home/sandra/Documents/matcha/eva_data/yksiksonad_2/yksiksonad_2_wav_resampled'):
        f.write(line + '\n')
    
    yksiklaused_txt_path = '/path/matcha/eva_data/yksiklaused/yksiklaused_txt'
    for line in create_training_set_row(yksiklaused_txt_path, '/home/sandra/Documents/matcha/eva_data/yksiklaused/yksiklaused_wav_resampled'):
        f.write(line + '\n')
```
